```python
import pandas as pd
```

### 1. Đọc và hiển thị toàn bộ dữ liệu

```python
path = f"./data/P4AI_BT1.csv"
db = pd.read_csv(path)
print(db.to_string())
```

```python
db.info()
```

```python
features = ["sepal.length", "sepal.width","petal.length","petal.width","variety"]
for item in features:
    print(db[item].value_counts())
```

```python
print(db.value_counts())
```

```python

print(db.isna().sum())

db1 = db
```

### 2. Điền các trường còn thiếu theo phương pháp trung bình với các giá trị số và theo phương pháp tần số cao nhất với các giá trị rời rạc

#### 2.1 Điền các trường giá trị số theo phương pháp giá trị trung bình

```python
db1['petal.length'].fillna(round(db['petal.length'].mean()),inplace = True)
db1['petal.width'].fillna(round(db['petal.width'].mean()),inplace = True)
db1['sepal.length'].fillna(round(db['sepal.length'].mean()), inplace=True)
db1['sepal.width'].fillna(round(db['sepal.width'].mean()), inplace=True)
db1
```

#### 2.2 Điền các trường rời rạc với phương pháp tần số cao nhất

```python
# print(db1['variety'])
db1['variety'].fillna(db1['variety'].mode()[0],inplace= True)
print(db1.to_string())
```

```python
print(db1.isna().sum())
```

#### 3. Hiển thị các bản ghi có sepal.width > 3 sepal.length > 5

```python

db_split = db1.loc[(db1['sepal.width'] > 3) & (db1['sepal.length'] > 5)]
print(db_split.to_string()) #
```

#### 4. Chuẩn hóa dữ liệu theo min max

```python
db2 = db1
def fill_min_max(df):
    max_column = df.max()
    min_column = df.min()
    df = (df - min_column)/(max_column - min_column)
    return df
db2['petal.length'] = fill_min_max(db2['petal.length'])
db2['petal.width'] = fill_min_max(db2['petal.width'])
db2['sepal.length'] = fill_min_max(db2['sepal.length'])
db2['sepal.width'] = fill_min_max(db2['sepal.width'])
```

```python
print(db2.to_string())
```

#### 5. Liên tục hóa thuộc tính variety

```python
arr = db2['variety'].unique()
for i in range(len(arr)):
    db2['variety'] = db2['variety'].replace(arr[i], i)
print(db2.to_string())
```

#### 6. Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
if 'variety' in db2.columns:
    X = db2.drop('variety', axis=1)
    y = db2['variety']
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"Mean Squared Error: {mse}")
    print(f"R-squared: {r2}")
    print("Coefficients:", model.coef_)
    print("Intercept:", model.intercept_)
```

#### 7. Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu

```python
sample_random = db2.sample(frac=0.5, replace=True, random_state=42)
print(len(sample_random))
print(sample_random.to_string())
```

#### 8. Rời rạc hóa thuộc tính sepal.length

```python
db2['sepal_length_bins'] = pd.cut(db2['sepal.length'], bins=4, 
                                  labels=["Low", "Medium", "High", "Very High"])
print(db2[['sepal.length', 'sepal_length_bins']].to_string())
```

#### 9. Phân cụm dữ liệu và loại bỏ ngoại lai 

```python
from sklearn.cluster import KMeans
features = ["sepal.length", "sepal.width"]
# ,"petal.length","petal.width"
X = db2[features]
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
print('Centers found by scikit-learn:')
print(kmeans.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
name = "sepal"

plt.scatter(db2['sepal.length'], db2['sepal.width'],
            c=kmeans.labels_, cmap='viridis')
plt.xlabel(f'sepal.length')
plt.ylabel(f'sepal.width')
plt.title('KMeans Clustering')


centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red',
            marker='x', s=100, label='Centers')

plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ biểu đồ hộp cho cột 'sepal.length'
plt.figure(figsize=(6, 4))
features = [f"{name}.length", f"{name}.width"]
sns.boxplot(x=f'{name}.length', data=db2[features])
plt.title(f'Box Plot - {name} Length')
plt.show()
```

```python
from scipy.stats import zscore

# Tính z-score cho cột 'sepal.length'
db2[f'{name}.length_zscore'] = zscore(db2[f'{name}.length'])
print(db2[f'{name}.length_zscore'].max())
print(db2[f'{name}.length_zscore'].min())
threshold = 2
outliers = db2[abs(db2[f'{name}.length_zscore']) > threshold]

# Loại bỏ các giá trị ngoại lai
X_cleaned = db2[abs(db2[f'{name}.length_zscore']) <= threshold]
print(len(X_cleaned))
print(X_cleaned.to_string())
```

```python
features = ["sepal.length", "sepal.width"]
# ,"petal.length","petal.width"
X = X_cleaned[features]
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
print('Centers found by scikit-learn:')
print(kmeans.cluster_centers_)
plt.scatter(X_cleaned['sepal.length'], X_cleaned['sepal.width'],
            c=kmeans.labels_, cmap='viridis')
plt.xlabel(f'sepal.length')
plt.ylabel(f'sepal.width')
plt.title('KMeans Clustering')


centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red',
            marker='x', s=100, label='Centers')

plt.legend()
plt.show()
```
